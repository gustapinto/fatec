```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
nlp_pt = spacy.load('pt_core_news_sm')
```

### Carregando textos externos

```python
from goose3 import Goose

url = 'https://en.wikipedia.org/wiki/Natural_language_processing'

# Instanciando o web scrapper
g = Goose()

# <Goose>.extract(...) -> Extrai os dados de uma url, trazendo as
# informações presentes na página
article = g.extract(url)

# <article>.title -> Título da página
print(article.title)

# <article>.cleaned_text -> Texto da página, sem muitos dos elementos
# HTML
print()
print(article.cleaned_text)

# <article>.authors -> Autores da página
print()
print(article.authors)
```

### Aplicando NLP ao texto extraído

```python
doc = nlp(article.cleaned_text)

# displacy.render(doc, style='ent', jupyter=True)
for entity in doc.ents:
    if entity.label_ == 'PERSON':
        print(entity)
```

### Montando nuvem de palavras

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud


cloud = WordCloud().generate(article.cleaned_text)

plt.figure(figsize=(8,8 ))
plt.imshow(cloud)
plt.axis('off')
```

```python
def preprocessing(sentence):
    sentence = (
        sentence.lower()
                .replace('.', ' ')
                .replace('[', ' ')
    )
    tokens = [t.text for t in nlp(sentence) if not (t.is_stop
                                               or t.like_num
                                               or t.is_punct
                                               or t.is_space
                                               or len(t) == 1)]
    return ' '.join(tokens)
```

```python
article_cleaned = preprocessing(article.cleaned_text)

print(len(article.cleaned_text), len(article_cleaned))
```

## Exercícios

### Exercício 1

Considerando a página da wikipedia sobre a cidade de Araras, exiba:

- o gráfico de NER proveniente da página
- uma relação apenas com os verbos que aparecem na página
- uma relação apenas com as organizações citadas na página
- as 20 palavras mais frequentes da página

```python
from goose3 import Goose
from nltk import FreqDist
from nltk.tokenize import word_tokenize


g = Goose()
url = 'https://pt.wikipedia.org/wiki/Araras_(S%C3%A3o_Paulo)'

article = g.extract(url)
doc = nlp_pt(article.cleaned_text)

print('NER:\n')
displacy.render(doc, style='ent', jupyter=True)

print('Verbos:\n')
for token in doc:
    if token.pos_ == 'VERB':
        print(token)

print('\nOrgs:\n')
for ent in doc.ents:
    if ent.label_ == 'ORG':
        print(ent)

tokens = word_tokenize(article.cleaned_text)
commom = FreqDist(tokens).most_common(20)

print('\nPalavras frequentes:\n')
for palavra, frequencia in commom:
    print(f'{palavra}-{frequencia}')
```

### Exercício 2

- Tendo como base o artigo da wikipedia sobre a cidade de Araras, exiba duas nuvens de palavras, sendo a primeira com o texto original e a segundo após um processamento para eliminar do texto elementos de pouco interesse.
- Exiba as 20 palavras mais frequentes depois do pre-processamento e compare os resultados com os alcançados no exercício 1.

```python
import matplotlib.pyplot as plt
from goose3 import Goose
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud


def print_word_cloud(text):
    cloud = WordCloud().generate(text)
    plt.figure(figsize=(8,8 ))
    plt.imshow(cloud)
    plt.axis('off')


def preprocessing(sentence):
    sentence = sentence.lower().replace('.', ' ').replace('[', ' ')
    tokens = [t.text for t in nlp_pt(sentence) if not (t.is_stop
                                               or t.like_num
                                               or t.is_punct
                                               or t.is_space
                                               or len(t) == 1)]
    return ' '.join(tokens)


def print_most_commom(text, n = 20):
    tokens = word_tokenize(text)
    commom = FreqDist(tokens).most_common(n)
    
    print('\nPalavras frequentes:\n')
    for palavra, frequencia in commom:
        print(f'{palavra}-{frequencia}')

g = Goose()
url = 'https://pt.wikipedia.org/wiki/Araras_(S%C3%A3o_Paulo)'
article = g.extract(url)

original_text = article.cleaned_text
processed_text = preprocessing(original_text)

print_word_cloud(original_text)
print_word_cloud(processed_text)
print_most_commom(original_text)
print_most_commom(processed_text)
```
